```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img=np.ones((400,400,3),dtype="uint8")*255
cv2.line(img,(50,50),(350,50),(255,0,0),3)
cv2_imshow(img)
```

```python
cv2.rectangle(img,(50,100),(355,300),(0,255,0),7)
cv2.rectangle(img,(200,100),(300,250),(0,0,255),-1)
cv2_imshow(img)
```

```python
cv2.circle(img,(200,100),50,(255,255,255),-9)
cv2_imshow(img)
```

```python
cv2.putText(img,"OpenCV",(50,380),cv2.FONT_HERSHEY_COMPLEX_SMALL,4,(25,100,70),1)
cv2_imshow(img)
```

```python
print("Pixel at (100,100)",img[100,100])
img[100,100]=[0,0,5]
img[101:110, 101:110] = [0, 255, 0]
cv2_imshow(img)
```
