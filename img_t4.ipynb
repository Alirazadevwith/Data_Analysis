```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread('download.webp')
print("Original shape",img.shape)
resized=cv2.resize(img,(300,300))
cv2_imshow(resized)
print("Resized shape",resized.shape)
```

```python
(h,w)=img.shape[:2]
center=(h//2,w//2)
M=cv2.getRotationMatrix2D(center,160,0.5)
rotate=cv2.warpAffine(img,M,(h,w))
cv2_imshow(rotate)
```

```python
f=cv2.flip(img,1)
cv2_imshow(f)
```

```python
f=cv2.flip(img,0)
cv2_imshow(f)
```

```python
f=cv2.flip(img,-1)
cv2_imshow(f)
```

```python
crp=img[50:200,10:400]
cv2_imshow(crp)
```
