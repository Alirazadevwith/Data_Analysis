```python
# convert the data-type of col rate

def handlerate(value):
  value=str(value).split('/')
  value=value[0]
  return float(value)
  df['rate']=df['rate'].apply(handlerate)
  print(df.head())
df.head(5)
```

```python
sns.set_style('darkgrid')
colors=['red','black','blue','orange']
sns.countplot(x=df['listed_in(type)'],palette=colors)
plt.xlabel("type of rest")
plt.ylabel('count')
plt.title("Amazing")
plt.show()
```

```python
group_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes': group_data})
plt.plot(result, c='green',marker='o')
plt.xlabel('Type of resturant', c='red',size=20)
plt.ylabel('Votes',c='red',size=20)
plt.xticks(rotation=45)
plt.show()
```

```python
df.isnull()
df.isnull().sum()
df['vote'].fillna(0,inplace=True)
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['rate'], bins=15)
plt.title("Histogram distribution")
plt.show()
```

```python
couple=df['approx_cost(for two people)']
sns.countplot(x=couple)
print(couple.mean())
```

```python
# which mode recieve max rating
plt.figure(figsize=(10,6))
sns.boxplot(x='online_order', y='rate', data=df)
```
