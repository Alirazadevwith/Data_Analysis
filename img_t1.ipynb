```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('download.webp')
if img is None:
  print("image not load")
else:
  print("img  found")
```

```python
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize=cv2.resize(img,(200,300))
cv2_imshow(resize)
```

```python
cv2.imwrite('new.jpg',resize)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
rgb=cv2.cvtColor(gray,cv2.COLOR_GRAY2BGR)
cv2_imshow(rgb)
```
